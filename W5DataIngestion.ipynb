**Data**<br>
We have taken some huge data (>2GB) here. Let's check the size of the data.

````python
'''
  Using the OS module's stat function to read the size of the file
'''
import pathlib
# calculate file size in KB, MB, GB
def convert_bytes(size):
    """ Convert bytes to KB, or MB or GB"""
    for x in ['bytes', 'KB', 'MB', 'GB', 'TB']:
        if size < 1024.0:
            return "%3.1f %s" % (size, x)
        size /= 1024.0

file_size = pathlib.Path('/content/drive/My Drive/Colab Notebooks/Books_rating.csv')
print("Size of the file is", file_size.stat().st_size, "bytes")
f_size = file_size.stat().st_size

x = convert_bytes(f_size)
print("Size of the file is {}".format(x))
````

**Converting Data into Text file and then zip file:**

````python
with open('/content/drive/My Drive/Colab Notebooks/Books_rating.csv', 'r') as f_in, open('Books_rating.txt', 'w') as f_out:
    content = f_in.read().replace(',', '|')
    f_out.write(content)
````
